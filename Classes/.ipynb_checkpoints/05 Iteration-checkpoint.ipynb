## 1. Multiple assignment

* As we have seen, we can make more than one assignment to the same variable.
* A new assignment makes an existing variable refer to a new value (and stop refering to the old value)

````python
x = 18
print("Before new assignment", x)
x = 20
print("After new assignment", x)
````

````python
a = 5
b = a
a = 3
print("A:", a,"B:", b)
````

## 2. Updating variables

One of the most commons forms of multiple assignment is an **update**, where the new value depends on the old one:

````python
#Adding one to a variable x
x = 2
x = x + 1
print(x)
````

````python
# When it's a persons' birtdhay
age = 18
age = age + 1
print(age)
````

````python
#countdown
countdown = 3
countdown = countdown - 1
print(countdown)
````

There's a form where we can update a variable without using the whole name:

````python
x = 2
x += 1
print(x)
````

````python
countdown = 3
countdown -= 1
print(countdown)
````

<font color=red> Updating a variable by adding 1 is called an **increment**; subtracting 1 is called **decrement.**</font>

## 3. The while statement

* Computers are often used to automate repetitive tasks.
* Repeating identical or similar tasks without making errors is easy for computers.
* We can read the **while** statement as if it were English:

````python
# Example of a while loop
n = 3
while (n > 0):
    print(n)
    n = n - 1
    
# Afuera del while
print("Done")
````

````python
# Print 5 times "Hello, World!"
n = 0
greetings = "Hello, World!"
while(n < 5):
    print(greetings)
    n += 1
````

<font color=red>It's important to increment or decrement the variable that will be inside the while condition. Otherwise, we will have an **INFINITE LOOP**</font>

````python
# Example of an infinite loop
n = 10
greetings = "Hello, World!"
while(n > 5):
    print(greetings)
    n += 1
````

### 3.1 break

* Sometimes we don't know if it's time to end a loop until we are half way through the body.
* In that case we can use the **break** statement to jump out of the loop

````python
# Using break

calif = [10, 8, 10, 9, -1]
i = 0
while(calif[i] > 0):
    print(calif[i])
    if(calif == -1):
        break
    i += 1
    
````

````python
print("Dame un valor")
input()
````

## 4. For statement

* First we will see a brief introduction of lists in order to use them in loops.
* A **list** can be defined as follows:

````python
animals = ["dog", "cat", "lion"]
````

````python
animals[3]
````

In Python we can use **for loops** to iterate over the items of any sequence (a list or a string), in the order that they appear in the sequence.

````python
for animal in animals:
    print(animal)
````

````python
# Iteration over a list of strings
for animal in animals:
    if(animal == "lion"):
        break
    print(animal)
````

<font color=red>Note the use of the reserved keywords **for** and **in.**</font>

````python
# Iteration over a list of int
calif = [8, 7, 9, 10]
for cal in calif:
    print(cal)
````

````python
# Iteration over a list of float
calif = [8.3, 7.2, 9.1, 10]
for c in calif:
    print(c)
````

````python
# Iteration over a string
longString = "This is a long string"
for letter in longString:
    print(letter)
````

### 4.1 The  <font color = blue;> range( ) </font> function

* If we need to iterate over a sequence of numbers, the built-in function <font color = blue;>range( )</font> comes in handy. 
* It generates arithmetic progressions

````python
#[0, 1, 2, 3, 4, 5)
for i in range(6):
    print(i)
````

````python
# Determine the initial and the end point
for x in range(1,6):
    print(x)
````

````python
# Determining the step in a range function
# range(inicial, final, step)
for num in range(1,21,2):
    print(num)
````

````python
# [21, 1)
for x in range(21, 0, -2):
    print(x)
````

````python
#Print the first 10 even numbers
for i in range(2,21,2):
    print(i)
````

````python
for i in range(21):
    if(i % 2 == 0):
        print(i)
````

## 5. Nested iteration

Let's look at the next example and see the importance of the use of a nested iteration

````python
# 1 * 1 = 1
# 1 * 2 = 2
# 1 * 3
# ...
# 1 * 12 = 12

# 2 * 1 = 1
# 2 * 2 = 2
# 2 * 3
# ...
# 2 * 12 = 12

# 12 * 1


#ciclo que itera sobre tablas
for i in range(1, 13):
    print("Tabla del", i)
    for j in range(1,13):
        #1, 2, 3, ... 12
        # i * j = (i * j)
        print(i, "*", j, "=", i *j )
    print("\n")





````

````python
# Multiplication tables example
for i in range(2, 13):
    print("Tabla del", i)
    # i = 2
    for j in range(2, 13):
        # j = 1
        # 2 * 1 = 2
        print(i, "*", j, "=", i*j)
    print("\n")
    
````

## 6. The zip function

To loop over two or more sequences at the same time, the entries can be paired with the **zip ( )** function

````python
stud_names = ['Juan', 'Paola', 'Arturo', 'Brenda']
stud_id = ['102155', '111423', '117423', '127138']
````

````python
for student, ID in zip(stud_names, stud_id):
    print(f'{student}: {ID}')
````
